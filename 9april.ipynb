Q3. How is Bayes' theorem used in practice?

Ans:Bayes' theorem is used in practice to calculate the probability of a cause given an observed effect, allowing for the updating of beliefs or predictions based on new information; this is particularly valuable in fields like medicine, where it can help interpret medical test results by considering the prior probability of a disease alongside the test accuracy, or in machine learning where it is used to build classifiers that continuously refine their predictions as they receive new data


Q4. What is the relationship between Bayes' theorem and conditional probability?

Ans:
Bayes theorem is used to determine conditional probability. When two events A and B are independent, P(A|B) = P(A) and P(B|A) = P(B)

Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?

Ans:
To choose the right Naive Bayes classifier for a problem, analyze the distribution of your features:
Gaussian Naive Bayes:
Use this when your features are continuous and follow a normal (Gaussian) distribution. This is the most common choice for continuous data like measurements or numerical values.
Multinomial Naive Bayes:
Select this when dealing with discrete data where features represent counts or frequencies, like word counts in text classification.
Bernoulli Naive Bayes:
Opt for this if your features are binary (true/false, yes/no) or if you want to treat your features as binary even if they have more values.
Key factors to consider when choosing a Naive Bayes classifier:
Data type:
Continuous: Gaussian Naive Bayes
Discrete (counts): Multinomial Naive Bayes
Binary: Bernoulli Naive Bayes
