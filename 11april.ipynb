Q1. What is an ensemble technique in machine learning?

Ans.An ensemble technique in machine learning refers to a method where multiple machine learning models are combined to produce a final prediction, aiming to achieve better accuracy than any single model alone by leveraging the strengths of each individual model and mitigating their weaknesses; common ensemble techniques include bagging, boosting, and stacking.
Key points about ensemble techniques:
Combining predictions:
Ensemble methods train multiple models on the same data, then combine their predictions through voting, averaging, or other strategies to generate a final output.
Diversity is key:
The effectiveness of an ensemble model often depends on the diversity of the individual models, meaning they should make different types of errors to complement each other.
Common applications:
Ensemble techniques are commonly used for both classification and regression tasks.
Popular ensemble methods:
Bagging (Bootstrap Aggregating):
Creates multiple models by training each on a random subset of the data, then combines their predictions by averaging.
Boosting:
Sequentially trains models, where each new model focuses on correcting the errors of the previous model.
Stacking:
Trains a meta-model to combine the predictions of multiple base models,

Q7. How is the confidence interval calculated using bootstrap?

Ans:To calculate a confidence interval using bootstrapping, you repeatedly take random samples (with replacement) from your original dataset, calculate the statistic of interest (like the mean) for each resample, and then use the percentiles of this "bootstrap distribution" to determine the confidence interval boundaries

taking a sample from the population, then repeatedly resampling from that sample with replacement, calculating the desired statistic for each resample, and finally analyzing the distribution of those calculated statistics to understand the variability of the original statistic

Q9. A researcher wants to estimate the mean height of a population of trees. They measure the height of a
sample of 50 trees and obtain a mean height of 15 meters and a standard deviation of 2 meters. Use
bootstrap to estimate the 95% confidence interval for the population mean height.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
sample_size = 50
sample_mean = 15  # mean height of the sample
sample_std = 2    # standard deviation of the sample
n_bootstrap = 10000  # number of bootstrap samples

# Generate the original sample data
np.random.seed(42)  # Set seed for reproducibility
sample_data = np.random.normal(loc=sample_mean, scale=sample_std, size=sample_size)

# Perform bootstrap
bootstrap_means = []
for _ in range(n_bootstrap):
    # Create a bootstrap sample by sampling with replacement
    bootstrap_sample = np.random.choice(sample_data, size=sample_size, replace=True)
    # Calculate the mean of the bootstrap sample
    bootstrap_means.append(np.mean(bootstrap_sample))

# Convert the bootstrap means list to a numpy array for easier indexing
bootstrap_means = np.array(bootstrap_means)

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)

# Output the results
print(f"Bootstrap 95% Confidence Interval for the population mean height: ({lower_bound:.2f}, {upper_bound:.2f})")

# Optionally, plot the distribution of bootstrap sample means
plt.hist(bootstrap_means, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(x=lower_bound, color='r', linestyle='dashed', linewidth=2, label=f'2.5th Percentile: {lower_bound:.2f}')
plt.axvline(x=upper_bound, color='r', linestyle='dashed', linewidth=2, label=f'97.5th Percentile: {upper_bound:.2f}')
plt.axvline(x=np.mean(bootstrap_means), color='b', linestyle='solid', linewidth=2, label='Bootstrap Mean')
plt.title("Bootstrap Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
